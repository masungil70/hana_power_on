```python
import pandas as pd 

stock = pd.Series(
    [50000, 51000, 49000, 53000],
    index=["Mon", "Tue", "Wed", "Thu"]
)

# 평균 주가
print("평균:", stock.mean())

# 최고가
print("최고:", stock.max())

# 상승한 날짜
print(stock[stock > 50000])
```

```python
import pandas as pd
import numpy as np

# STEP 1: 데이터 로딩
df = pd.read_csv("student_100.csv")

# STEP 2: 구조 확인
print(df.info())

# STEP 3: 결측치 처리
df["나이"] = df["나이"].fillna(df["나이"].mean())
df["점수"] = df["점수"].fillna(0)

# STEP 4: 데이터 선택
df = df[["이름", "점수", "학과"]]

# STEP 5: 데이터 가공
df["등급"] = df["점수"].apply(
    lambda x: "A" if x >= 90 else "B"
)

# STEP 6: 분석
result = df.groupby("학과")["점수"].mean()

print(result)
```

```python
import pandas as pd

# 데이터 생성
data = {
    '부서': ['개발', '개발', '인사', '인사', '개발'],
    '나이': [29, 31, 35, 28, 29],
    '급여': [300, 450, 250, 280, 500]
}

df = pd.DataFrame(data)

# 여러 컬럼 기준 그룹화
result = df.groupby(['부서', '나이'])['급여'].mean()

print(result)
```

```python
import pandas as pd

# 데이터 생성
data = {
    '부서': ['개발', '개발', '인사'],
    '급여': [300, 500, 250]
}

df = pd.DataFrame(data)

# 그룹화 후 평균 계산
result = (
    df.groupby('부서')['급여']
      .mean()
)

print(result)
print("--------------------")


# 그룹화 후 평균 계산
result = (
    df.groupby('부서')['급여']
      .mean()
      .reset_index()
)

print(result)
```

```python
import pandas as pd

# 거래 데이터 생성
data = {
    '카테고리': ['식비', '쇼핑', '식비', '교통', '쇼핑', '교통'],
    '금액': [12000, 50000, 15000, 7000, 30000, 9000]
}

df = pd.DataFrame(data)

# 카테고리별 통계 계산
result = (
    df.groupby('카테고리')['금액']
      .agg(['sum', 'mean', 'max'])
      .sort_values(by='sum', ascending=False)
)

print(result)

result = (
    df.groupby('카테고리')['금액']
      .agg(['sum', 'mean', 'max'])
      .reset_index()
      .sort_values(by='sum', ascending=False)
)

print(result)
```

```python
customer = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '이름': ['김철수', '이영희', '박민수']
})

order = pd.DataFrame({
    '고객아이디': [1, 1, 2],
    '주문금액': [10000, 20000, 15000]
})

# 고객ID 기준으로 INNER JOIN 수행
result = pd.merge(
    customer,
    order,
    left_on='고객ID',
    right_on='고객아이디',
    how='inner'
)

print(result)
```
